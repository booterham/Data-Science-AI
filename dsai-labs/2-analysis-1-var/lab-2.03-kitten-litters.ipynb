## Exercise 3 - Kitten litters

A vet keeps a record of the number of kittens in litters produced by cats in his practice. The code block below contains the data as a frequency table.

- First put the data into one list of individual observations: [1, 1, 2, 2, ... 5, 5, 6, 6, 6, 6, 7, 7, 8]  
- Visualise the data set using all suitable chart types that you know of.
- Calculate:
    - the median size of a litter
    - the mean size of a litter
    - the lower quartile
    - the upper quartile

```python
kittens = pd.DataFrame.from_dict({
    'Number in litter': [1, 2, 3,  4, 5, 6, 7, 8],
    'Frequency':        [2, 4, 7, 11, 8, 4, 2, 1]})
kittens
```

```python
litters = np.repeat(kittens['Number in litter'], kittens['Frequency'])
```

```python
sns.catplot(data=litters, kind='count', color=bakermiller)
```

```python
# Create a list of individual observations based on frequency
sns.displot(litters, color=bakermiller, binwidth=1)
plt.xlabel('Number in litter')
plt.ylabel('Count')
```

```python
sns.kdeplot(data=litters, color=bakermiller, fill=True, bw_adjust=0.5)
```

```python
sns.violinplot(data=litters, color=bakermiller);
```

```python
from statsmodels.stats.weightstats import DescrStatsW

median = litters.quantile(0.5)
mean = litters.mean()
lower_quartile = litters.quantile(0.25)
upper_quartile = litters.quantile(0.75)
print(f"Median: {median}, Mean: {mean}, Lower Quartile: {lower_quartile}, Upper Quartile: {upper_quartile}")
```
