## Vraag 4 [3 pt]
Een hockeyteam wint met een kans van 0.6 en verliest met een kans van 0.3.
Het team speelt drie wedstrijden in het weekend. Bereken de kans dat het team:

1. [1 pt] alle drie de wedstrijden wint.
2. [1 pt] minstens twee keer wint en niet verliest.
3. [1 pt] één wedstrijd wint, één verliest en één gelijk speelt (in willekeurige volgorde).

```python
W = 0.6
V = 0.3
G = 0.1

print(f"De kans dat het team drie wedstrijden wint is {pow(W, 3)*100}%")
print(f"De kans dat het team minstens twee wedstrijden wint en niet verliest {(pow(W, 2) * (1-0.3))*100}%")
WVG = W * V * G
print(f"De kans dat het team één wedstrijd wint, één verliest en één gelijk speelt is {(WVG)*100}% als het in die volgorde is")
print("Mogelijke volgorden voor gegeven combinatie:")
perm = list(itertools.permutations(['A', 'B', 'C'], 3))
print(perm)
total_perm = len(perm)
print(f"Totaal aantal mogelijke volgorden is {total_perm}")
print(f"Dus totale kans met willekeurige volgorde is: {total_perm * WVG*100}%")
```

### Antwoord vraag 4

Stel:
- $W$ het team wint.
- $V$ het team verliest.
- $G$ het team speelt gelijkspel.

3 elementen kiezen uit een set van 3 zonder herhaling is gelijk aan $3! = 6$

1. $$\mathcal P(\text{team wint 3 wedstrijden}) = \mathcal P(W) \cdot \mathcal P(W) \cdot \mathcal P(W) = 0.6^3 \approx 21.60\%$$
2. $$\mathcal P(\text{team wint minstens 2 keer en verliest niet}) = \mathcal P(W) \cdot \mathcal P(W) \cdot \mathcal P(\overline V)\\= 0.6^2 \cdot (1 - 0.3)\\=25.2\%$$
3. $$\mathcal P(\text{team wint een keer, verliest een keer en speelt een keer gelijkspel, in willekeurige volgorde})\\=3! \cdot \mathcal P(W)\cdot \mathcal P(V) \cdot \mathcal P(G)\\\approx 10.80\%$$


## Vraag 5 [3 pt]

Aan een bepaalde instelling voor hoger onderwijs ontvangt iedereen met een eindscore van meer dan 85 in het afstudeerjaar een premie van 1000 EUR en iedereen met een eindscore tussen 80 en 85 ontvangt 500 EUR. Alle anderen krijgen niets. De eindscores zijn normaal verdeeld met een gemiddelde van 63 en een standaardafwijking van 12.

Stel dat er 5000 studenten afstuderen, bereken dan de totale kost voor het verstrekken van de premies voor deze instelling.

Je kan dit berekenen in een drietal stappen. Vermeld duidelijk de gebruikte stappen.

```python
threshold_pink = 85
premie_pink = 1000
threshold_green = 80
premie_green = 500

mu = 63
s = 12

n = 5000

# normaalverdeling bekijken
dist_x = np.linspace(mu - 4 * s, mu + 4 * s, num=201)
dist_y = stats.norm.pdf(dist_x, mu, s)
plt.plot(dist_x, dist_y, color=bakermiller)

plt.fill_between(dist_x, 0, dist_y, where=(dist_x >= threshold_pink), color='pink')
plt.fill_between(dist_x, 0, dist_y, where=(dist_x >= threshold_green) & (dist_x < threshold_pink), color='yellowgreen')
plt.axvline(threshold_pink, color="orange", label="85");
plt.axvline(threshold_green, color="skyblue", label="80");
plt.legend()
```

```python
green_percentage = stats.norm.cdf(85, loc=mu, scale=s) - stats.norm.cdf(80, loc=mu, scale=s)
green_students = round(green_percentage * n)
green_price = green_students * premie_green
print(f"De proportie studenten die tussen de 80 en 85 behaalt is: {green_percentage*100}%.")
print(f"Dit komt overeen met {green_students} studenten.")
print(f"Dit komt overeen met €{green_price:.2f}.")


pink_percentage = stats.norm.sf(85, loc=mu, scale=s)
pink_students = round(pink_percentage * n)
pink_price = pink_students * premie_pink
print(f"\nDe proportie studenten die minstens 85 behaalt is: {pink_percentage*100}%.")
print(f"Dit komt overeen met {pink_students} studenten.")
print(f"Dit komt overeen met €{pink_price:.2f}.")

total_price = pink_price + green_price
print(f"\nHet totaalbedrag is €{total_price:.2f}")
```

## Vraag 6 [5 pt]

De gemiddelde dagelijkse hoeveelheid schroot van een bepaald fabricageproces is 25.5 kg met een standaardafwijking van 1.6 kg. Er wordt geprobeerd het proces aan te passen in een poging deze hoeveelheid te verminderen.Tijdens een proefperiode werden de onderstaande kilo's schroot geproduceerd. Door de aard van de wijziging werd de dagelijkse variabiliteit van de hoeveelheid schroot niet veranderd. De normale verdeling is van toepassing. Een eerste blik op de cijfers suggereert dat de wijziging effectief is in het verminderen van het schrootniveau. Bevestigt een significantietest dit op 1%-niveau?

1. [1 pt] Welke hypothesetoets ga je toepassen om deze onderzoeksvraag te beantwoorden? Wees zo specifiek mogelijk!
2. [1 pt] Formuleer de nulhypothese en de alternatieve hypothese
3. [1 pt] Bereken de gepaste toetsingsgrootheid (teststatistiek) voor deze toets. Geef het symbool en de waarde
4. [1 pt] Bereken de p-waarde
5. [1 pt] Trek een besluit op basis van de vorige stap en formuleer een antwoord op de onderzoeksvraag.


```python


schroot = [25.70, 24.98, 24.30, 24.43, 23.21, 23.74, 24.03, 26.76, 23.83, 24.02,  25.34, 26.12, 26.46, 23.61, 24.76, 26.92, 25.05, 24.59, 24.47, 25.37, 26.28, 22.50, 24.13, 24.63, 25.08, 24.59, 20.90, 26.70, 23.07, 19.94, 24.30, 22.53, 24.09, 23.63, 23.84]

mu = 25.5
s = 1.6
n = len(schroot)
m_sample = np.mean(schroot)
alpha = 0.01
SE = s / np.sqrt(n)

p = stats.norm.cdf(m_sample, loc=mu, scale=SE)

print(f"De teststatistiek is {m_sample}")
print(f"De p-waarde is: {p*100}%")
```

```python
# normaalverdeling bekijken
dist_x = np.linspace(mu - 5 * SE, mu + 5 * SE, num=201)
dist_y = stats.norm.pdf(dist_x, mu, SE)
plt.plot(dist_x, dist_y, color=bakermiller)

plt.fill_between(dist_x, 0, dist_y, where=(dist_x >= threshold_pink), color='pink')
plt.fill_between(dist_x, 0, dist_y, where=(dist_x < m_sample), color='yellowgreen')
plt.axvline(m_sample, color="orange", label="85");
```

### Antwoord vraag 6

1. Een one-sided, left-tailed z-toets.
2. $H_0$: Het gemiddelde van het dagelijks geproduceerd schroot tijdens de proefperiode is niet significant lager. Dus $\mu = 25.5k$g
   
   $H_1$: Het gemiddelde van het dagelijks geproduceerd schroot tijdens de proefperiode is significant lager. Dus $\mu < 25.5k$g
3. De teststatistiek $\overline x \approx 24.39714285714286$
4. De kans dat een waarde uit de originele verdeling nog kleiner is dan het nieuwe gemiddelde is $p \approx 0.002272515338531605% $. 
5. De $p$-waarde is kleiner dan $\alpha = 0.01$, dus we kunnen met zekerheid de nulhypothese verwerpen in het voordeeel van de alternatieve hypothese. Het nieuwe gemiddelde van de hoeveelheid schroot dat geproduceerd wordt per dag, is significant lager dan het vorige gemiddelde.

 ## Vraag 10 [4 pt]

Een bepaald type kleine landbouwmachines wordt geproduceerd door vier verschillende toeleveringsbedrijven. Het aandeel van de machines die gerepareerd moesten worden in het eerste jaar na verkoop aan de landbouwers is als volgt

| Bedrijf | Totaal aantal machines | Aandeel dat gerepareerd moet worden |
| ------ | ------ | ------ |
| A | 145 | 0.1034 |
| B | 140 | 0.0429 |
| C | 120 | 0.0333 |
| D | 105 | 0.1143 |

Is de verdeling met betrekking tot het vereisen en niet vereisen van reparaties onafhankelijk van het toeleveringsbedrijf? Gebruik het 5% significantieniveau.  

[1 pt] Welke hypothesetoets ga je toepassen om deze onderzoeksvraag te beantwoorden? Wees zo specifiek mogelijk!  
[1 pt] Formuleer de nulhypothese en de alternatieve hypothese  
[1 pt] Bereken de p-waarde  
[1 pt] Trek een besluit op basis van de vorige stap en formuleer een antwoord op de onderzoeksvraag.  

```python
dfmachines = pd.DataFrame(data={
    'bedrijf': ['A', 'B', 'C', 'D'],
    'totaal aantal verkochte machines': [145, 140, 120, 105],
    'percentage dat gerepareerd moet worden': [0.10, 0.05, 0.03, 0.12]
})

alpha = 0.05

observed = pd.crosstab(dfmachines['bedrijf'], dfmachines['percentage dat gerepareerd moet worden'])
chi2, p, df, expected = stats.chi2_contingency(observed)

print(f"p-waarde: {p}")
```

### Antwoord vraag 10

1. De $\chi^2$-toets voor onafhankelijkheid
2. $H_0$ De verdeling met betrekking tot het al dan niet vereisen van reparaties is onafhankelijk van het toeleveringsbedrijf.
   
   $H_1$ De verdeling met betrekking tot het al dan niet vereisen van reparaties is afhankelijk van het toeleveringsbedrijf.
3. De $p$-waarde is een stuk groter dan $0.01$, namelijk: $p\approx 0.21330930508341653$.
4. De nulhypothese dat de verhouding te repareren machines onafhankelijk is van het soort bedrijf kan niet verworpen worden. We besluiten dat de verdeling met betrekking tot het al dan niet vereisen van reparatie onafhenkelijk is van het toeleveringsbedrijf.
