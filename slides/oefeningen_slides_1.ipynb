# Functions



```python
import scipy.stats as stats
```

# Exercices

```python
prob = stats.norm.cdf(3, 0, 1)
print(f"The probability P(X < 3) = {prob:.8f}")
```

```python
lesthanX = stats.norm.sf(-2, 0, 1)
morethanX = stats.norm.cdf(6.5, 0, 1)
print(f"The probability P(X > -2 and X < 6.5) = {lesthanX * morethanX:.8f}")
```

```python
x = stats.norm.ppf(0.80, 0, 1)
print(f"For x = {x:.4f}, P(X < x) = 80%")
```
